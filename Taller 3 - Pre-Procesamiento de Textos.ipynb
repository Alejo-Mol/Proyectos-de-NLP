```python
with open("UNA_SOLITARIA_VOZ_HUMANA.txt", encoding='utf-8') as archivo:
    data = archivo.read()
```

```python
#Vamos a convertirlas todoas en minusculas 
data= data.lower()
#Vamos a convertir solo a palabras
data = re.sub(r"[\W\d_]+"," ",data)
#Vamos a quitar los espacios
data = data.strip()
#Vamos a hacer tokenizacion
data = data.split()
data
```

# Punto 2:

- `[8pts]` Quitar palabras vacias
- `[9pts]` ¿Cuáles son las 10 palabras no vacias más usadas?

```python
from nltk.corpus import stopwords
stopwords_sp = stopwords.words("spanish")
```

```python
#Vamos a quitar palabras vacias 
data =  [palabra for palabra in data if palabra not in stopwords_sp]
```

```python
from collections import Counter
```

```python
cnt = Counter()
for i in data:
    cnt[i] += 1

```

```python
Counter(data).most_common(10)
```

# Punto 3:

- `[8pts]` Stemming del documento
- `[9pts]` ¿Cuáles son las 10 raíces más usadas?

```python
from nltk.stem.snowball import SnowballStemmer
englishStemmer=SnowballStemmer("english")
spanishStemmer=SnowballStemmer("spanish")
```

```python
stem_data=[]
for i in data:
     stem_data = stem_data + [spanishStemmer.stem(i)]
```

```python
cnt = Counter()
for i in stem_data:
    cnt[i] += 1

```

```python
Counter(stem_data).most_common(10)
```
